```python
import numpy as np
import matplotlib.pyplot as plt

x_values = np.linspace(0, 20, 100)
y_values = np.sin(x_values)

plt.plot(x_values, y_values)
plt.show()

print(x_values)
print(y_values)
```

```python
x = np.linspace(0, 5, 200)
y1 = 2 * x
y2 = x ** 2
y3 = np.log(x)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

```python
x = np.linspace(0, 10, 11)
y1 = x ** 2
y2 = x ** 3
y3 = x ** 4
print(x)
plt.plot(x, y1,  color='r', linestyle='dashed', label='x^2')
plt.plot(x, y2, color='b', linestyle='dotted', label='x^3')
plt.plot(x, y3, color='green', linestyle='dashdot', label='x^4')
plt.legend(loc='upper left') 
plt.show()
```

```python
x = np.linspace(0, 5, 200)
y1 = x ** 2
y2 = x ** 3
y3 = x ** 4
plt.subplot(311) # subplot(xyz), where x = the number of graphs, y = graph in each line, z = order
plt.plot(x, y1, 'r-')
plt.subplot(312)
plt.plot(x, y2, 'g--')
plt.subplot(313)
plt.plot(x, y3, color='y', linestyle='dotted') 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 180, 6
x = mu + sigma * np.random.randn(10000)

plt.hist(x, 1000, density=True, facecolor="blue")
plt.xlabel("height")
plt.ylabel("probability")
plt.title("Height of students")
plt.text(160, 0.125, "mu=172, sigma=4")
plt.axis([155, 220, 0, 0.2])
plt.grid(True)
```

```python
bob = ( 90 , 67 , 87 , 76 )
charles = ( 99 , 99 , 90 , 95 )
daniel = ( 40 , 60 , 76 , 22 )
skills = ( "Python" , "Java" , "Networking" , "Machine Learning")

width = 0.2
index = np.arange( 4 )
plt.bar(index, bob,
width =width, label = "Bob" )
plt.bar(index + width, charles,
width =width, label = "Charles" )
plt.bar(index + width * 2 , daniel,
width =width, label = "Daniel" )

plt.xticks(index + width, skills)
plt.ylim( 0 , 120 )
plt.title( "IT Skill Levels" )
plt.ylabel( "Skill Level" )
plt.xlabel( "IT Skill" )
plt.legend()
```

```python
# Pie chart
labels = ( 'Pizza' , 'Hamburger' , 'Sushi' , 'Spagheti' )
values = ( 47 , 23 , 20 , 10 )
plt.pie(values, labels =labels,
autopct = "%.2f%%" , shadow = True )
plt.title( "Favorite Food" )
plt.show()
```
