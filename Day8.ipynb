```python
import numpy as np
import matplotlib.pyplot as plt

x_values = np.linspace(0, 20, 100)
y_values = np.sin(x_values)

plt.plot(x_values, y_values)
plt.show()

print(x_values)
print(y_values)
```

```python
x = np.linspace(0, 5, 200)
y1 = 2 * x
y2 = x ** 2
y3 = np.log(x)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

```python
x = np.linspace(0, 10, 11)
y1 = x ** 2
y2 = x ** 3
y3 = x ** 4
print(x)
plt.plot(x, y1,  color='r', linestyle='dashed', label='x^2')
plt.plot(x, y2, color='b', linestyle='dotted', label='x^3')
plt.plot(x, y3, color='green', linestyle='dashdot', label='x^4')
plt.legend(loc='upper left') 
plt.show()
```

```python
x = np.linspace(0, 5, 200)
y1 = x ** 2
y2 = x ** 3
y3 = x ** 4
plt.subplot(311) # subplot(xyz), where x = the number of graphs, y = graph in each line, z = order
plt.plot(x, y1, 'r-')
plt.subplot(312)
plt.plot(x, y2, 'g--')
plt.subplot(313)
plt.plot(x, y3, color='y', linestyle='dotted') 
plt.show()
```
